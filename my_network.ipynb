```python
import networkx as nx
%load_ext autoreload
%autoreload 2
import physquirrel as sq

G = nx.Graph()
G.add_node(1, color='blue')
G.add_edge(1, 2, weight=3.5)

print(G.graph)         # Graph attributes
print(G.nodes[1])
print(G.edges[2,1])
print(G.nodes.data())  # Node attributes
print(G.edges.data())  # Edge attributes
print(G.edges)
```

```python
G1 = sq.DirectedNetwork([(1,2), (2,3), (3,4), (4,1)])
print(G1.is_tree())
print(G1.root_node())
```

```python
G2 = sq.SemiDirectedNetwork( leaves = {3,5,7}, directed_edges = {(2,6), (4,6)}, incoming_graph_data= [(1,2), (2,3), (1,4), (4,5), (4,6), (2,6), (6,7)])
print(G2.is_tree())
print(G2.edges)
print(G2.nodes)
print(G2.reticulation_nodes())
in_v = G2.indegree(6)
deg_v = G2.degree(6)
print(in_v, deg_v)
```
